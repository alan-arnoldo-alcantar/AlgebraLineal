## **Álgebra Lineal**

### 🎬 Una historia de cine (y de álgebra lineal)

Imagina que vas al cine con tus amigos un viernes por la tarde. No hay spoilers todavía, la película promete, y lo único que te preocupa es una cosa:

👉 ***¿Qué combo de comida te conviene más?***

Miras el menú arriba del mostrador. Hay varios combos, cada uno con su combinación de palomitas, sodas y nachos:

|**Combo**                 |🍿 Palomitas |🥤 Sodas |🧀 Nachos |💵 Precio |
|--------------------------|---------------|----------|------------|--------|
|**👩🏼‍🦰 Combo individual**  |      1 🍿     |   2 🥤   |   1 🧀    | $180   |
|**💑 Combo pareja**      |      2 🍿     |   1 🥤   |   2 🧀    | $240   |
|**🤤 Combo antojad@**    |      1 🍿     |   1 🥤   |   3 🧀    | $210   |
|**👨‍👩‍👧‍👦 Combo familiar**    |      1 🍿     |   1 🥤   |   3 🧀    | $210   |

Tu amigo te dice:
> — “¿Y si mejor compramos todo por separado? Tal vez sale más barato.”

Pero... ¡espera!
Ningún precio individual está a la vista. Solo aparecen los combos.

Entonces se te prende el foco 💡:
> “Si supiéramos cuánto cuesta cada cosa por separado, podríamos comparar y decidir mejor.”

### 🧠 Lo interesante
Este no es un problema de números raros ni de fórmulas locas.

Es un problema de lógica. De **relaciones simples** entre cantidades.

Y hay algo que damos por hecho sin pensarlo mucho, pero que es fundamental:

- ✅ El precio de una palomita es siempre el mismo, sin importar en qué combo venga.
- ✅ Lo mismo con la soda y los nachos.
- ❌ No hay descuentos secretos.

Si en un combo hay el doble de sodas, eso suma el doble de precio.
Si un combo trae más nachos, simplemente cuesta más.
Las reglas son claras, constantes y proporcionales.
Eso es lo que hace a este problema lineal.

### 📚 Y eso, sin decirlo, ya es álgebra lineal
Lo que estás intentando hacer es resolver un **sistema lineal**:

- Conoces el total de cada combinación.
- Quieres descubrir los valores individuales que lo componen.

Y lo haces con **sentido común**, sin necesidad de saber lo que es un vector, una matriz o una transformación lineal (todavía).

### 🌍 Lo mejor es que este tipo de razonamiento está en todos lados
El mismo tipo de lógica se usa para:

- Calcular el costo total de un pedido en una tienda en línea.
- Balancear fuerzas en una estructura de ingeniería.
- Estimar ingredientes en una receta ajustada para más personas.
- Modelar redes eléctricas.
- Entrenar modelos de machine learning.
- Simular personajes en videojuegos 3D.

> Así que sí: cuando estás frente a la pantalla del cine, tratando de descifrar cuánto valen las palomitas, ya estás **pensando como alguien que hace álgebra lineal**.
Solo falta que aprendas las herramientas para resolverlo con facilidad, incluso cuando el sistema se vuelve más grande.

Y eso es justo lo que vamos a explorar en estar notas.



## **Vectores**

> Una idea, muchas formas de imaginarlos!!

Los vectores son una de esas ideas que aparecen en todas partes: en la física, la computación, las finanzas, la ingeniería… ¡hasta en los videojuegos! Pero lo interesante es que no hay una sola forma de imaginarlos. Dependiendo del contexto, puedes verlos como:

* Una flecha que apunta en cierta dirección y tiene cierto tamaño.
* Una lista de números que describe algo con precisión.

Y lo mejor es que ambas formas están diciendo lo mismo, solo que en diferentes “idiomas”.

### 🧑🏽‍🔬👩🏻‍🚀👷🏾 **Visión de los físicos/ingenieros**

**🚗 ¿Cómo le explicas a un coche autónomo hacia dónde ir?**

Imagina que estás en un coche con conducción automática, como un Tesla o el auto de Google con GPS. Tú no le dices “ve por allá”, porque el coche no entiende qué es “allá”. Lo que sí entiende es: gira el volante 30 grados a la derecha y avanza 20 metros. Este tipo de instrucciones le dice al coche exactamente **hacia dónde debe ir** (la dirección del movimiento) y **qué tan lejos** debe avanzar (la magnitud).

Los ingenieros que programan estos autos usan este tipo de instrucciones todo el tiempo para planear rutas, evitar obstáculos y moverse con precisión. Imagina estas instrucciones como "*flechas*" imaginarias, que sin ellas el coche no tendría idea de cómo llegar a ningún lado.

**⛵️ ¿Cómo se mueve un barco si solo tiene viento?**

Imagina que estás en un velero en medio del mar. No hay motor, no hay remos… solo el viento. Y aunque no lo veas, puedes sentir cómo el viento **empuja** al barco.

Pero el viento no solo sopla “*fuerte”*. También sopla en **cierta dirección**. No es lo mismo que sople directo desde el norte a que venga desde un ángulo, o desde atrás del barco.

Cuando el viento actúa sobre el velero, lo está empujando con **cierta fuerza** (qué tan fuerte sopla) y **hacia un lugar específico** (la dirección en la que sopla). Es como si una mano invisible empujara al barco con una "*flecha*": la flecha apunta hacia donde el viento lo mueve, y su tamaño depende de qué tan fuerte lo está empujando.

Quienes diseñan velas, cascos o timones —como los ingenieros navales o civiles especializados en estructuras— necesitan entender exactamente cuánto empuja el viento y en qué dirección lo hace, para que el barco no pierda el equilibrio ni se desvíe.

**🎮 Mover a tu personaje también es dar instrucciones con dirección y cantidad**

Piensa en un videojuego como Minecraft, Fortnite o cualquier juego donde controlas a un personaje en un mundo 3D o 2D.

Cuando mueves el stick para moverte hacia adelante o hacia un lado y presionas el botón para correr, lo que en realidad estás haciendo es decirle al juego: muéveme en **esta dirección** y con esta **velocidad**. Y cuando combinas varios movimientos (como correr hacia adelante y saltar), estás dando una instrucción más compleja: muévete hacia arriba y adelante al mismo tiempo, con cierta velocidad.

Aunque tú solo veas al personaje moverse, el juego está usando números para calcular **hacia dónde debe ir** (la dirección del movimiento) y **cuánto** debe moverse (la distancia, la velocidad, o la fuerza).

Es como si cada movimiento fuera una "*flecha*" digital que apunta hacia donde el personaje debe ir, y que tiene cierto largo dependiendo de qué tan rápido o lejos debe moverse.

Los programadores y diseñadores de videojuegos usan estas "flechas" para todo: para mover personajes, lanzar objetos, calcular saltos, hacer que el viento empuje algo en el mapa o que una bala siga cierta trayectoria.

### 👩🏻‍⚕️🧑🏾‍🏫👩🏼‍💻 **Visión de los informaticos/sociales/salud**

**🎸🎷🎹 ¿Cómo sabe Spotify qué música te gusta?**

Imagina que usas Spotify todos los días. A veces escuchas reggaetón para motivarte, otras veces rock cuando estás con amigos, y pop cuando estás en el transporte. La app va registrando todo eso, pero no lo guarda como “Alan escucha música variada”.

Lo que hace Spotify es ir apuntando cuántas canciones de cada tipo escuchas. Y esa información se organiza en una lista que, por ejemplo, puede verse así:

| Género musical |  🧔🏾‍♂️ Alan |  👩🏻‍🦰 Geni |
|----------------|-----------|-----------|
| Reggaetón      |    10     |     3     |
| Rock           |     3     |     7     |
| Pop            |     5     |     9     |
| Clásica        |     0     |     2     |
| Electrónica    |     2     |     6     |

Cada número está en un lugar específico y tiene un significado (para Alan):

* El primero es reggaetón (10 canciones)
* El segundo es rock (3 canciones)
* El tercero es pop (5 canciones)
* Y así con los demás.

Esa **lista ordenada** le permite a Spotify comparar lo que te gusta con lo que le gusta a otras personas parecidas a ti. Y así, cuando te recomienda una nueva canción, no lo hace al azar: lo hace basándose en esa especie de “huella musical” que fuiste dejando con tus hábitos.

**🤔 ¿Cómo se usan los números para entender lo que piensan las personas?**

Imagina que estás participando en una encuesta sobre temas sociales. Te preguntan qué tanto estás de acuerdo con frases como:

* “Confío en el gobierno”
* “Las redes sociales afectan mi bienestar”
* “La educación pública es de calidad”
* “Los influencers tienen impacto real”
* “La tecnología mejora la vida”

Y tú respondes del 1 al 5 según lo que piensas. Las respuestas de varias personas quedan en listas como estas:

| Pregunta                         | Persona A 🥷🏻 | Persona B 🥷🏻 |
|-----------------------------------------|-----------|-----------|
| Confío en el gobierno                   |     2     |     4     |
| Las redes sociales afectan mi bienestar |     4     |     3     |
| La educación pública es de calidad      |     3     |     2     |
| Los influencers tienen impacto real     |     5     |     5     |
| La tecnología mejora la vida            |     4     |     5     |


Parece solo una **lista de números…** pero en realidad cada uno de ellos representa una opinión sobre un tema distinto, y el orden en el que están **importa mucho**. La primera posición corresponde a la primera pregunta, la segunda a la segunda, y así sucesivamente.

Los investigadores pueden juntar miles de estas listas para ver qué temas generan más acuerdo, cuáles dividen a la gente, o qué grupo de personas piensa de forma parecida. Es una manera de convertir opiniones en algo que se puede estudiar, comparar y entender mejor.

**💊🧬 Tu cuerpo también puede contarse con números**

Vas al médico a hacerte un chequeo. Te toman el pulso, la presión, la oxigenación y la temperatura. Al final te dan un pequeño reporte con algo así:

| Análisis clínico          | Paciente 🤒 | Intervalo saludable ❤️‍🩹 |
|---------------------------|------------|--------------------------|
| Glucosa (mg/dL)           |    105     | 70 – 99                 |
| Presión arterial (mmHg)   |  130/85     | Menor a 120/80          |
| Colesterol total (mg/dL)  |    190     | Menor a 200             |
| Frecuencia cardiaca (bpm) |     75     | 60 – 100                |
| Hemoglobina (g/dL)        |    13.5     | 13.5 – 17.5 (hombres)   |

Y cada valor en esa **lista** tiene un significado dependiendo de su **posición**: la primera entrada es la glucosa, la segunda la presión, etc. , Lo importante es **dónde está cada una**: si las cambiaras de lugar, podrían significar cosas completamente distintas.

Los doctores usan este tipo de listas para entender el estado general del cuerpo, pero también para ver si hay algo raro o fuera de lo normal. En hospitales grandes, sistemas automáticos revisan miles de estas listas por día buscando señales de alerta.


### **😥 ¿Flechas o listas? ¡Las dos cosas! 🤩**

Hasta ahora hemos visto dos formas de entender los vectores que usamos en muchos campos:
por un lado, como flechas con dirección y tamaño, y por otro, como listas de números que representan algo. A veces tiene más sentido imaginarlo como una flecha, otras veces como una lista.
Lo importante es entender que **ambas representaciones hablan del mismo tipo de objeto matemático**.

> Puede parecer que estamos hablando de cosas distintas, pero en realidad... ¡son dos formas de ver lo mismo!

A veces los vectores parecen flechas.
Otras veces, solo listas de números.

Y está bien.
No se trata de escoger una sola forma de verlos.
Lo valioso es poder cambiar de perspectiva cuando lo necesites.

Verlos como flechas te ayuda a imaginar movimientos, fuerzas, trayectorias.
Verlos como listas te permite trabajar con datos, cálculos y simulaciones.

> Saber moverte entre esas dos ideas es lo que realmente importa.
Esa flexibilidad es lo que hace al álgebra lineal tan útil en el mundo real.

### 🧮📓 **Visión de los matematicos**

**¿Y los matemáticos qué dicen?**

Hasta ahora hemos visto vectores como flechas o como listas.
Pero cuando les preguntas a los matemáticos, ellos te dan otra forma de verlos.

Para ellos, un **vector** no tiene por qué ser una flecha ni una lista concreta.
Un vector es simplemente una **"cosa"** que cumple ciertas reglas.

* Si puedes **sumarlo** con otros vectores
* y **multiplicarlo** por un número normal comun y corriente…

si esas operaciones tienen sentido y se comportan bien,
entonces probablemente estés trabajando con un **vector**.

Esto puede sonar abstracto, pero tiene una gran ventaja:
permite que la idea de vector se use en muchos lugares,
incluso donde no hay flechas ni listas,
como funciones, imágenes, sonidos, o hasta movimientos en videojuegos.

La matemática busca encontrar lo que todas esas cosas tienen en común,
y eso es lo que hace posible construir herramientas tan poderosas.

#### ⬆️➡️=🏹 **Suma de vectores:** ***“Cuando las flechas hacen equipo”***

Sumar vectores es tan simple como sumar cada posición de una lista con la misma posición de la otra.

Es como si tuvieras dos listas de gastos mensuales y sumaras “renta con renta”, “comida con comida”, “transporte con transporte”… al final obtienes una nueva lista con los totales.

En la visión de flechas, es como poner una flecha justo al final de otra y dibujar la línea que va desde el inicio de la primera hasta la punta de la segunda.

> Sumar vectores: combina la información de dos listas o une dos flechas.

####  **Multiplicación por un escalar:** ***“Estirando, encogiendo… y a veces volteando flechas”***

Aquí no mezclamos vectores, sino que tomamos un vector y lo “estiramos” o “encogemos” multiplicando cada uno de sus elementos de la lista que representa el vector por ese númmero.

Si pensamos en flechas, multiplicar por un número mayor que 1 hace que la flecha crezca; si el número está entre 0 y 1, la flecha se acorta; y si el número es negativo, además de cambiar su tamaño, también apunta en la dirección contraria.

> Multiplicar por un número: cambia el tamaño (y a veces la dirección) del vector.

## **Vectores base, combinacion lineal y espacio generado**

* Construir un vector a partir de la suma de los vectores base i y j escalados dando origen a la combinacion lineal
* Espacio generado por la combinacion lineal de dos vectores (cuales son todos los posibles vectores que puedes constuir usando suma y multiplicacion entre dos vectores dados)
* Conjunto de vectores independientes (por ejemplo los vetores base) construyen todo el espacio donde viven
* Las bases de un espacio vectorial es un conjunto de vectores linealmente independiente que generan el espacio vectorial completo
